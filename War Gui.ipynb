```python
import numpy as np
import random as rand

class War(object):
    def __init__(self):
        self.cards = {}
        self.refs = []
        for c in ['h','s','c','d']:
            for i in range(13):
                ref = c+str(i+1)
                self.refs = np.append(self.refs, ref)
                self.cards[ref] = Image.open("cards/"+ref+".bmp")
        self.players = [0, 1, 2, 3]
            

    def deal(self):
        hand = list(self.refs)
        rand.shuffle(hand)
        hand = np.resize(hand, (4,13))
        self.p1 = hand[0]
        self.p2 = hand[1]
        self.p3 = hand[2]
        self.p4 = hand[3]
        self.hand = [self.p1, self.p2, self.p3, self.p4]
        return

    def draw(self):
        play = []
        spot = []
        win = []
        for i, array in enumerate(self.hand):
            if len(array) == 0:
                del self.hand[i]
                del self.players[i]
            else:
                play = np.append(play, array[0])
                self.hand[i] = np.delete(array, 0)

        for i, card in enumerate(play):
            win = np.append(win, int(play[i][1:]))
        spots = np.where(win == max(win))
        spots = spots[0]
        if len(spots) == 1:
            self.hand[spots[0]] = np.append(self.hand[spots[0]], play)
            self.hand = self.hand[spots[0]:]+self.hand[:spots[0]]
            self.players = self.players[spots[0]:]+self.players[:spots[0]]
        else:
            self.war(spots, winnings = play)
        

        return play

    def war(self, spots, winnings=[]):
        winner = []
        for i in spots:
            winnings = np.append(winnings, self.hand[i][:4])
            winner = np.append(winner, int(winnings[-1][1:]))
            self.hand[i] = self.hand[i][4:]
        newspot = np.where(winner == max(winner))[0]
        if len(newspot) > 1:
            self.war(newspot, winnings = winnings)
        else:
            win = spots[newspot[0]]
            self.hand[win] = np.append(self.hand[win], winnings)
            self.hand = self.hand[win:]+self.hand[:win]
            self.players = self.players[win:]+self.players[:win]
            
```

```python
from tkinter import *
from PIL import ImageTk, Image
import time

class Application(Frame):
    def deal(self):
        self.game.deal()
        self.p1d.configure(image=self.back)
        self.p2d.configure(image=self.back)
        self.p3d.configure(image=self.back)
        self.p4d.configure(image=self.back)
        
    def play(self):
        players = self.game.players
        trick = self.game.draw()
        for i, player in enumerate(players):
            self.ps[player].configure(image = self.cards[trick[i]])
            self.master.update()
            time.sleep(.5)
        for p in [0,1,2,3]:
            if p not in players:
                self.ps[p].configure(image = self.blank)
            
        
        

    def createWidgets(self):
        self.QUIT = Button(self)
        self.QUIT["text"] = "QUIT"
        self.QUIT["fg"]   = "red"
        self.QUIT["command"] =  self.quit

        self.QUIT.grid(row = 0, column = 0)
        
        self.Deal = Button(self)
        self.Deal["text"] = "Deal"
        self.Deal["command"] = self.deal
        
        self.Deal.grid(row = 0, column = 1)
        
        self.p1p = Label(self, image = self.blank)
        self.p2p = Label(self, image = self.blank)
        self.p3p = Label(self, image = self.blank)
        self.p4p = Label(self, image = self.blank)
        self.p1p.grid(row = 4, column=2)
        self.p2p.grid(row = 3, column=1)
        self.p3p.grid(row = 2, column=2)
        self.p4p.grid(row = 3, column=3)
        self.ps = [self.p1p, self.p2p, self.p3p, self.p4p]
        
        self.p1d = Button(self, image = self.blank)
        self.p1d.grid(row = 5, column=2)
        self.p1d["text"] = 'PLAY'
        self.p1d["command"] = self.play
        
        self.p2d = Label(self, image = self.blank)
        self.p2d.grid(row = 3, column=0)
        self.p3d = Label(self, image = self.blank)
        self.p3d.grid(row = 1, column=2)
        self.p4d = Label(self, image = self.blank)
        self.p4d.grid(row = 3, column=4)
        
        

        


    def __init__(self, master=None):
        Frame.__init__(self, master)
        self.game = War()
        self.cards = self.game.cards
        blank = Image.open('cards/blank.bmp')
        self.blank = ImageTk.PhotoImage(blank)
        back = Image.open('cards/back.bmp')
        self.back = ImageTk.PhotoImage(back)
        for card in self.cards:
            self.cards[card] = ImageTk.PhotoImage(self.cards[card])
        
        self.pack()
        self.createWidgets()


root = Tk()
app = Application(master=root)
app.mainloop()
root.destroy()
```
